# Rocket Path Visualization

Visualizes the rocket's travel path (x, z coordinates) recorded in the `position_data.csv` file.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

## 1. Load CSV File

Loads the `position_data.csv` file generated from the Unity project.

```python
# Specify the absolute path to the CSV file.
csv_file_path = '/home/perelman/Easy_robot/My project/position_data.csv'

# Check if the CSV file exists.
if not os.path.exists(csv_file_path):
    print(f"Could not find CSV file at '{csv_file_path}'")
    print("Please run the Unity project first to generate the 'position_data.csv' file.")
else:
    with open(csv_file_path, 'r') as f:
        lines = f.readlines()

    data_lines = lines
    # Check if the first line is a header, and if so, skip it.
    if lines:
        first_line_cols = lines[0].strip().split(',')
        try:
            # If it's a data row, the distance (index 1) or time (index 2) should be convertible to a number.
            float(first_line_cols[1])
            float(first_line_cols[2])
        except (ValueError, IndexError):
            # If conversion to a number fails, treat it as a header and exclude it from the data.
            print("Header row detected, skipping.")
            data_lines = lines[1:]

    # Split each line by comma to create a list of data.
    data = [line.strip().split(',') for line in data_lines if line.strip()]

    if not data:
        print("No data in the CSV file (excluding header).")
        # Create an empty DataFrame to prevent errors in subsequent cells.
        df = pd.DataFrame()
    else:
        # Create a DataFrame using the list of data.
        df = pd.DataFrame(data)

        # Dynamically generate column names based on the longest row.
        max_cols = df.shape[1]
        col_names = []
        if max_cols > 0: col_names.append('TestID')
        if max_cols > 1: col_names.append('Distance')
        if max_cols > 2: col_names.append('Time')
        
        if max_cols > 3:
            for i in range(3, max_cols):
                point_index = (i - 3) // 2
                axis = 'x' if (i - 3) % 2 == 0 else 'z'
                col_names.append(f'P{point_index}_{axis}')
                
        df.columns = col_names

        print("CSV file loaded successfully!")
        print(f"Found {len(df)} run records.")
        display(df.head())
```

## 2. Define Path Visualization Function

Defines a function to select a specific run record from the DataFrame and plot its path.

```python
def visualize_path(run_index):
    """
    Visualizes the travel path for a given run index.
    run_index: The index of the run record to visualize (starting from 0).
    """
    if 'df' not in globals() or df.empty:
        print("Data has not been loaded. Please run the previous cell first.")
        return

    if run_index < 0 or run_index >= len(df):
        print(f"Invalid index. Please enter a value between 0 and {len(df) - 1}.")
        return

    # Extract path data from the selected row and convert non-numeric values to NaN.
    path_data_series = df.iloc[run_index, 3:]
    path_data_numeric = pd.to_numeric(path_data_series, errors='coerce')
    
    # Remove NaN values to keep only valid path data.
    path_data = path_data_numeric.dropna().values

    if len(path_data) < 2:
        print("Not enough path data to plot.")
        return

    # Adjust data length to ensure only complete (x, z) pairs are used.
    num_points = len(path_data)
    if num_points % 2 != 0:
        # Remove the last incomplete x-coordinate.
        path_data = path_data[:-1]

    # Separate x and z coordinates.
    x_coords = path_data[0::2]
    z_coords = path_data[1::2]

    # Check if there are any valid coordinates.
    if len(x_coords) == 0:
        print("No valid path coordinates found.")
        return

    # Plot the path.
    plt.figure(figsize=(10, 10))
    plt.plot(x_coords, z_coords, marker='.', linestyle='-', label=f'Run {run_index}')
    
    # Highlight the start and end points.
    plt.plot(x_coords[0], z_coords[0], 'go', markersize=10, label='Start')
    plt.plot(x_coords[-1], z_coords[-1], 'ro', markersize=10, label='End')

    plt.title(f'Path Visualization for Run #{run_index}')
    plt.xlabel('X Coordinate')
    plt.ylabel('Z Coordinate')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')  # Set equal scaling for X and Y axes to show the path without distortion.
    plt.show()
```

## 3. Run Visualization

Execute the `visualize_path` function by passing the number (index) of the run record you want to see.

```python
# Select the index of the run record to visualize (starting from 0).
# For example, to see the first run record, enter 0.
visualize_path(0)
```
